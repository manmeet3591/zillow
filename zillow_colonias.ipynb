```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('County_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv')
# Filter for Hidalgo County in Texas
df_filtered = df[(df['RegionName'] == 'Hidalgo County') & (df['StateName'] == 'TX')]

# Extract the time series data
time_series = df_filtered.iloc[0, 10:]
time_series.index = pd.to_datetime(time_series.index)
time_series = pd.to_numeric(time_series, errors='coerce')

# Plot
plt.figure(figsize=(12, 6))
plt.plot(time_series.index, time_series.values, label='ZHVI - Hidalgo County, TX')
plt.title('Zillow Home Value Index (ZHVI)\nMcAllen-Edinburg-Mission Metro')
plt.xlabel('Date')
plt.ylabel('ZHVI ($)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
